## 3. Data Cleaning

```python
#Defining the function to clean up all the reviews 
import string
from nltk.corpus import stopwords

def message_cleaning(message):
    test_punc_removed = [char for char in message if char not in string.punctuation]
    test_punc_removed_join = ''.join(test_punc_removed)
    test_punc_removed_join_clean = [word for word in test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return test_punc_removed_join_clean

#Applying the new function to reviews
reviews_df_clean = reviews_df['verified_reviews'].apply(message_cleaning)
```

```python
#Checking clean reviews
print(reviews_df_clean.iloc[5])
```

```python
#Checking original reviews
print(reviews_df['verified_reviews'].iloc[5])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#Performing tokenization
vectorizer = CountVectorizer(analyzer = message_cleaning)
reviews_countvectorizer = vectorizer.fit_transform(reviews_df['verified_reviews'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(reviews_countvectorizer.toarray())
```

```python
#Checking shape of reviews_countvectorizer
reviews_countvectorizer.shape
```

```python
reviews = pd.DataFrame(reviews_countvectorizer.toarray())
X = reviews 
y = reviews_df['feedback']
```

```python
y
```

## 4. Model Building & Evaluation

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 100)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

#Predicting test set results
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict_test))
```

```python
#Building a Logistic regression Model & compring results

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)

print(classification_report(y_test, y_pred))
```

We can see the Logistic Regression Model performed a little better than Naive Bayes Model. It was able to achieve an weighted avg of 0.92 while the Naive Bayes Model it was only 0.91
